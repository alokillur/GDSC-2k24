```python
# def prediction():
#     list = [str(p1._Country), str(p1._Locality), str(p1._Industry), str(p1._Gender), str(p1._PartyPerson), str(p1._CriticalRisk), str(p1._Year), str(p1._Day), str(p1._Month), str(p1._Description)]
#     Data = '' 
#     for x in list:
#         Data += x
#         Data += ' '

#     tensor_x = text_to_tensor(data)(Data, 3173, unk_token='__UNK__', verbose=False)

#     from tensorflow.keras.preprocessing.sequence import pad_sequences
#     maxlen = 100

#     padding = [0]*(maxlen - len(tensor_x))

#     X = tensor_x + padding
#     X = np.array(X)
#     X.shape = (1, 100)

#     y = model.predict(X, batch_size=1)

#     output = y.argmax(axis=1)[0]
#     return output


def text_to_tensor(text, vocab_size, max_length, unk_token='__UNK__', verbose=False):
    # Tokenize the text (assuming space-separated tokens)
    tokens = text.split()

    # Map tokens to indices using a vocabulary
    word_to_index = {}
    for i, token in enumerate(tokens):
        if token not in word_to_index:
            word_to_index[token] = len(word_to_index) + 1  # Start indexing from 1
    if unk_token not in word_to_index:
        word_to_index[unk_token] = len(word_to_index) + 1  # Assign an index to the unknown token
    if verbose:
        print("Word to index mapping:", word_to_index)

    # Convert tokens to indices and handle out-of-vocabulary tokens
    tensor = [word_to_index.get(token, word_to_index[unk_token]) for token in tokens]

    # Pad or truncate the tensor to the desired length
    if len(tensor) < max_length:
        tensor += [0] * (max_length - len(tensor))  # Pad with zeros
    else:
        tensor = tensor[:max_length]  # Truncate to max_length
    return tensor


def prediction():
    # Assuming p1._Country, p1._Locality, ..., p1._Description are strings
    data_list = [str(p1._Country), str(p1._Locality), str(p1._Industry), str(p1._Gender), str(p1._PartyPerson), str(p1._CriticalRisk), str(p1._Year), str(p1._Day), str(p1._Month), str(p1._Description)]
    data = ' '.join(data_list)

    # Use the text_to_tensor function
    tensor_x = text_to_tensor(data, 3173, unk_token='__UNK__', verbose=False)

    # Rest of your code...
    from tensorflow.keras.preprocessing.sequence import pad_sequences
    maxlen = 100

    padding = [0]*(maxlen - len(tensor_x))

    X = tensor_x + padding
    X = np.array(X)
    X.shape = (1, 100)

    y = model.predict(X, batch_size=1)

    output = y.argmax(axis=1)[0]
    return output

# Example usage
output_prediction = prediction()
print(output_prediction)  # Print or use the output as needed
```
